## Números Exponentes Enteros

*Alfred North Whitehead*

Las notaciones matemáticas evolucionan con el tiempo. Por ejemplo, lo que ahora escribimos como $x^2$ una vez se escribió como $xx$. Si bien es posible leer $xx$ de un vistazo, $xxxxxxxxxxx$ es más complicado. Para simplificar, se adoptó una notación abreviada:

**Definición (Exponentes Enteros no Negativos):** Para cualquier número entero no negativo $n$, definimos

$$x^n = \underbrace{x \cdot x \cdots x}_{n \text{ veces}}$$

Esta notación tiene propiedades algebraicas importantes que podemos descubrir simplemente experimentando. Por ejemplo,

$$x^5 \cdot x^{11} = \underbrace{x \cdot x \cdots x}_{5 \text{ veces}} \cdot \underbrace{x \cdot x \cdots x}_{11 \text{ veces}} = \underbrace{x \cdot x \cdots x}_{16 \text{ veces}} = x^{16}$$

De aquí, podemos deducir un principio más general: $x^n \cdot x^m = x^{n+m}$ para cualquier número entero no negativo $n$ y $m$.

Esta es la primera de cinco reglas básicas de los exponentes. Las otras cuatro son igualmente naturales y fáciles de entender. Las he enumerado a continuación:

**Reglas de los Exponentes**

1.  $x^n \cdot x^m = x^{n+m}$
2.  $\frac{x^n}{x^m} = x^{n-m}$
3.  $(x^n)^m = x^{nm}$
4.  $(xy)^n = x^n y^n$
5.  $\left(\frac{x}{y}\right)^n = \frac{x^n}{y^n}$

Memoriza estas cinco reglas, pero si alguna vez las olvidas, puedes recuperarlas fácilmente escribiendo un ejemplo numérico específico en un papel.


## Reglas de los Exponentes

| Regla | Expresión Matemática | Ejemplo | Explicación |
|---|---|---|---|
| 1 | $x^m \cdot x^n = x^{m+n}$ | $x^5 \cdot x^3 = x^8$ | Al multiplicar potencias de la misma base, se suman los exponentes. Esto es como sumar el número total de veces que se multiplica la base. Por ejemplo, $x^5$ significa multiplicar x cinco veces y $x^3$ significa multiplicar x tres veces. Al multiplicar ambas expresiones, obtenemos x multiplicado ocho veces, que es equivalente a $x^8$. |
| 2 | $\frac{x^m}{x^n} = x^{m-n}$ | $\frac{x^5}{x^3} = x^2$ | Al dividir potencias de la misma base, se restan los exponentes. Esto se debe a que al dividir, estamos cancelando los factores comunes. En el ejemplo, podemos cancelar tres factores x tanto en el numerador como en el denominador, quedando solo dos factores x en el numerador. |
| 3 | $(x^m)^n = x^{mn}$ | $(x^2)^3 = x^6$ | Cuando elevamos una potencia a otra potencia, se multiplican los exponentes. Esto significa que estamos elevando la base a la potencia resultante de multiplicar los exponentes. En el ejemplo, estamos elevando $x^2$ al cubo, lo que equivale a multiplicar $x^2$ por sí mismo tres veces. |
| 4 | $(xy)^n = x^n y^n$ | $(xy)^3 = x^3y^3$ | Al elevar un producto a una potencia, elevamos cada factor del producto a esa potencia. Esto se debe a que estamos multiplicando el producto por sí mismo n veces. Cada factor del producto se multiplica por sí mismo n veces. |
| 5 | $\left(\frac{x}{y}\right)^n = \frac{x^n}{y^n}$ | $\left(\frac{x}{y}\right)^3 = \frac{x^3}{y^3}$ | Al elevar un cociente a una potencia, elevamos tanto el numerador como el denominador a esa potencia. Esto es similar a la regla anterior, pero aplicado a una división. |


## Exponentes Negativos y Cero


Comencemos considerando la posibilidad de elevar algo a la potencia de 0. ¿Qué podría significar $x^0$? Nuestra definición de la sección anterior no aplica, ya que 0 no es un número entero positivo. Sin embargo, se podría argumentar que, dado que para números enteros positivos $n$, $x^n$ es el producto de multiplicar $n$ copias de $x$, la expresión $x^0$ debería ser lo que obtenemos al multiplicar cero copias de $x$. Bueno, cero copias de cualquier cosa es... ¡nada! Entonces, ¿$x^0$ debe ser igual a 0?

¡Incorrecto! Si insistimos en preservar las reglas de los exponentes, entonces tenemos que aceptar que, si la expresión $x^0$ debe tener algún significado, entonces $x^0 = 1$. Aquí, nos vemos obligados a aceptar que las reglas de los exponentes deben prevalecer sobre nuestra intuición.

¿Qué sugiere esta meticulosa elección de $2^0 = 1$? ¡Nada! Nuestra intuición (que insiste en que $x^0$ debería ser 0) choca con las reglas de los exponentes (que insisten en que $x^0 = 1$). Algo tiene que ceder. Los matemáticos han decidido que las reglas de los exponentes deben prevalecer porque pueden demostrar su validez en situaciones mucho más complejas, donde nuestra intuición es bastante inútil.

Entonces, ¿cómo podemos darle un significado preciso a $x^{-n}$? Más generalmente, ¿qué significa elevar algo a la potencia de $-n$? Significa elevar algo a la potencia de $n$ y tomar su recíproco (en cualquier orden).

$$x^{-n} = \frac{1}{x^n}$$

Un caso especial es elevar algo a la potencia de -1. Esto simplemente significa tomar el recíproco de ese número.

## Demostración de $x^0 = 1$

Utilizando las reglas de los exponentes, podemos demostrar que $x^0 = 1$:

$$x^0 = x^{1-1} = \frac{x^1}{x^1} = \frac{x}{x} = 1$$

## Demostración de la propiedad de los exponentes negativos

**Teorema:** Para cualquier número real $x$ distinto de cero y cualquier número entero $n$, se cumple que:

$$x^{-n} = \frac{1}{x^n}$$

**Demostración:**

$$x^{-n} = x^{0-n} = \frac{x^0}{x^n} = \frac{1}{x^n}$$

**Y así, hemos demostrado que:**

$$x^{-n} = \frac{1}{x^n}$$

**Ejemplos:**

* $2^{-3} = \frac{1}{2^3} = \frac{1}{8}$
* $\left(\frac{2}{3}\right)^{-2} = \left(\frac{3}{2}\right)^2 = \frac{9}{4}$
* $(-2)^{-10} = \left(\frac{1}{-2}\right)^{10} = \frac{1}{(-2)^{10}} = \frac{1}{1024}$

**Definiciones (Exponentes Cero y Negativos):**

Para cualquier $x$ no nulo, hacemos las siguientes definiciones:

* $x^0 = 1$
* $x^{-n} = \frac{1}{x^n}$

En particular, elevar algo a la potencia de -1 es lo mismo que tomar su recíproco.
