# Identidad Fundamental

$$ \sin^2(x) + \cos^2(x) = 1 $$

Sabemos que en un círculo unitario, la relación entre los lados del triángulo rectángulo y el radio nos da que el cuadrado de las funciones seno y coseno sumando 1.

## Desarrollo de la Identidad

### 1. **Definiciones de Seno y Coseno**

Las funciones seno y coseno se definen en el contexto de un triángulo rectángulo. Para un ángulo \(x\):

- **Seno**: Es la razón entre el cateto opuesto y la hipotenusa.
  
$$
\sin(x) = \frac{\text{cateto opuesto}}{\text{hipotenusa}}
$$

- **Coseno**: Es la razón entre el cateto adyacente y la hipotenusa.
  
$$
\cos(x) = \frac{\text{cateto adyacente}}{\text{hipotenusa}}
$$

### 2. **Teorema de Pitágoras**

En un triángulo rectángulo, el teorema de Pitágoras nos dice que:

$$
\text{hipotenusa}^2 = \text{cateto opuesto}^2 + \text{cateto adyacente}^2
$$

Si llamamos \(h\) a la hipotenusa, \(a\) al cateto adyacente y \(b\) al cateto opuesto, podemos expresar esto como:

$$
h^2 = a^2 + b^2
$$

### 3. **Sustituyendo Seno y Coseno**

Usando las definiciones de seno y coseno:

- $(b = h \cdot \sin(x))$
- $(a = h \cdot \cos(x))$

Sustituyendo en el teorema de Pitágoras:

$$
h^2 = (h \cdot \cos(x))^2 + (h \cdot \sin(x))^2
$$

Esto se simplifica a:

$$
h^2 = h^2 \cdot \cos^2(x) + h^2 \cdot \sin^2(x)
$$

### 4. **Dividiendo Entre (h^2)**

Si $(h^2 \neq 0)$, podemos dividir ambos lados de la ecuación por $(h^2)$:

$$
1 = \cos^2(x) + \sin^2(x)
$$

Reordenando, obtenemos:

$$
\sin^2(x) + \cos^2(x) = 1
$$

## Conclusión

Esta identidad es fundamental en trigonometría y se utiliza en múltiples áreas de la matemática y la física. Se aplica en la resolución de ecuaciones trigonométricas y en la simplificación de expresiones trigonométricas complejas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el dominio
x = np.linspace(0, 2 * np.pi, 400)

# Definimos las funciones
sin2 = np.sin(x)**2
cos2 = np.cos(x)**2
identity = sin2 + cos2

# Graficamos
plt.plot(x, sin2, label=r'$\sin^2(x)$', color='r')
plt.plot(x, cos2, label=r'$\cos^2(x)$', color='b')
plt.plot(x, identity, label=r'$\sin^2(x) + \cos^2(x) = 1$', color='g', linestyle='--')

plt.title('Identidad $\sin^2(x) + \cos^2(x) = 1$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.axhline(1, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el círculo unitario
theta = np.linspace(0, 2 * np.pi, 400)
x = np.cos(theta)
y = np.sin(theta)

# Graficamos el círculo unitario
plt.plot(x, y, label='Círculo unitario')

# Marcamos el radio y un punto en el círculo
plt.plot([0, np.cos(np.pi / 4)], [0, np.sin(np.pi / 4)], 'r-', label='Radio')
plt.plot(np.cos(np.pi / 4), np.sin(np.pi / 4), 'ro', label=r'$(\cos(\theta), \sin(\theta))$')

# Ajustes de la gráfica
plt.gca().set_aspect('equal')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Círculo Unitario y la Identidad Fundamental')
plt.xlabel('x = cos(θ)')
plt.ylabel('y = sin(θ)')
plt.legend()
plt.grid(True)
plt.show()
```

# Identidad de la Tangente

$$ \tan(x) = \frac{\sin(x)}{\cos(x)} $$

La tangente se define como la razón entre el seno y el coseno. Esta función tiene asíntotas verticales en los puntos donde el coseno se anula, es decir, en $x = \frac{\pi}{2} + n\pi$.


```python
# Definimos las funciones

tan = np.tan(x)
sin = np.sin(x)
cos = np.cos(x)

# Graficamos
plt.plot(x, tan, label=r'$\tan(x)$', color='g')
plt.plot(x, sin, label=r'$\sin(x)$', color='r')
plt.plot(x, cos, label=r'$\cos(x)$', color='b')

# Ajustes de la gráfica
plt.title(r'Identidad $\tan(x) = \frac{\sin(x)}{\cos(x)}$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(np.pi/2, color='gray', linestyle='--')
plt.axvline(3*np.pi/2, color='gray', linestyle='--')
plt.ylim([-10, 10])
plt.legend()
plt.grid(True)
plt.show()
```

# Identidad de Ángulo Doble

$$ \sin(2x) = 2\sin(x)\cos(x) $$

Podemos demostrar esta identidad usando las fórmulas de suma de ángulos. Además, gráficamente se puede ver que la amplitud de $\sin(2x)$ es mayor que la de $\sin(x)$, y la frecuencia también es el doble.


```python
# Definimos la función
sin_2x = np.sin(2 * x)
sin_x_cos_x = 2 * np.sin(x) * np.cos(x)

# Graficamos
plt.plot(x, sin_2x, label=r'$\sin(2x)$', color='r')
plt.plot(x, sin_x_cos_x, label=r'$2\sin(x)\cos(x)$', color='b', linestyle='--')

plt.title(r'Identidad $\sin(2x) = 2\sin(x)\cos(x)$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.axhline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

# Límite Trigonométrico Importante

$$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$

A medida que $x$ se acerca a 0, la razón $\frac{\sin(x)}{x}$ se aproxima a 1. Esto se puede demostrar analíticamente, pero aquí lo vamos a ilustrar gráficamente.


```python
# Definimos el dominio más pequeño alrededor de 0
x_lim = np.linspace(-0.1, 0.1, 400)
sin_over_x = np.sin(x_lim) / x_lim

# Graficamos
plt.plot(x_lim, sin_over_x, label=r'$\frac{\sin(x)}{x}$', color='purple')
plt.axhline(1, color='gray', linestyle='--')

plt.title(r'Límite $\lim_{x \to 0} \frac{\sin(x)}{x} = 1$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.grid(True)
plt.legend()
plt.show()
```
