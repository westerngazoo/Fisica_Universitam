## La Propiedad Distributiva

La **propiedad distributiva** (o más precisamente, "la propiedad distributiva de la multiplicación sobre la suma") es el hecho de que multiplicar algo por una *suma* equivale a multiplicar ese mismo algo por *cada término individual* en la suma y luego sumar los resultados. Por ejemplo, la propiedad distributiva nos dice que

$$5(7 + 10) = (5 \cdot 7) + (5 \cdot 10)$$

y

$$(b + c + d)a = ab + ac + ad.$$

Cuando describimos estas operaciones en palabras, hablamos de "distribuir el 5" (o la a) sobre la suma. La resta es solo un tipo especial de suma (la suma de un negativo), por lo que también podemos distribuir la multiplicación sobre la resta. Así, la propiedad distributiva garantiza que

$$(x - y - w)z = xz - yz - wz$$

y

$$2a(a - b + c) = 2a^2 - 2ab + 2ac.$$

La propiedad distributiva es la base sobre la que se construye gran parte del álgebra, como verás en las siguientes páginas. También justifica algunas técnicas simples de aritmética mental, como verás en los ejercicios a continuación.

# FOILed Again
# (En Alabanza a la Distribución, Parte 1)

Al principio de tu primer curso de álgebra, aprendiste a multiplicar dos "binomios". Es decir, aprendiste a expandir $(a + b)(c + d)$ en la forma $ac + ad + bc + bd$. La mayoría de los profesores de álgebra resumen los pasos en este proceso con el acrónimo FOIL (Primero, Exterior, Interior, Último). El acrónimo es tan común que se ha convertido en un verbo, como en "cuando hacemos FOIL, obtenemos..."

Todos los estudiantes de álgebra saben cómo "FOIL", pero sorprendentemente pocos saben por qué funciona. Es una pena, ya que la explicación es muy sencilla. No implica nada más que la propiedad distributiva. La idea clave es que podemos, cuando es conveniente, pensar en un binomio como una sola entidad a distribuir. Para enfatizar esta idea, pondré un binomio en un cuadro gris cuando quiera destacar su naturaleza unitaria. Observa cuidadosamente:

$$(a + b)(c + d) = (a + b)(c + d)$$  
$$= (a + b)c + (a + b)d \quad \text{(distribuyendo la "caja", $(a + b)$)}$$  
$$= ac + bc + ad + bd \quad \text{(distribuyendo el $c$, y también el $d$)}$$

Así, detrás de escena, la misteriosa operación FOIL es solo una forma abreviada de llevar a cabo la propiedad distributiva varias veces. Si entiendes esto, puedes averiguar fácilmente cómo multiplicar dos trinomios, tres binomios, etc., sin tener que esperar a que alguien te enseñe nuevos acrónimos.


## Factorización de Polinomios
(En Alabanza a la Distribución, Parte 2)

“Factorizar” no es más que la distribución en reversa (o “desdistribución”, como la llamé en el Ejercicio 9).

Por ejemplo, ¿por qué es cierto que $3x^2 + 6x - 18 = 3(x^2 + 2x - 6)$? Si lees esa ecuación de derecha a izquierda, verás por qué: es la buena propiedad distributiva que justifica ese signo de igualdad.

Quizás has encontrado (y olvidado) la identidad de “diferencia de cuadrados”, $a^2 - b^2 = (a - b)(a + b)$. Leyéndolo de izquierda a derecha, es algo misterioso. ¿Qué hace que esta extraña fórmula de factorización sea cierta? ¿Por qué deberíamos creerlo? Bueno, si lo leemos de derecha a izquierda, vemos cómo podríamos probar que es cierto.

Haremos esto ahora, poniendo un binomio en un cuadro gris cuando lo pensemos como un bloque.

**Afirmación.** $a^2 - b^2 = (a - b)(a + b)$ para todo $a$ y $b$.

**Demostración.** 
$(a - b)(a + b) = (a - b)(a + b)$  
$\quad = (a - b)a + (a - b)b$ (distribuyendo el $a - b$)  
$\quad = a^2 - ab + ab - b^2$ (distribuyendo el $a$ y también el $b$)  
$\quad = a^2 - b^2$, como se afirmó. 

Nota cómo la prueba se reduce a tres aplicaciones de la propiedad distributiva – y nada más. No es exagerado decir que la fórmula de diferencia de cuadrados se sostiene gracias a la propiedad distributiva.

De hecho, la fórmula de diferencia de cuadrados es solo la punta del iceberg de la factorización. Cada factorización de polinomios puede justificarse "leyendo de derecha a izquierda", multiplicando los factores y verificando que el resultado es el polinomio original. Como hemos visto, multiplicar esos factores es, simplemente, una cuestión de distribuir repetidamente, así que toda la temática de la factorización de polinomios está justificada por la humilde propiedad distributiva. No hay nada profundo aquí.

La mayoría de los libros de álgebra hacen que parecer que factorizar un polinomio es un procedimiento complicado. No lo es. Casi cualquier polinomio que necesites factorizar a mano se rinde a uno (o alguna combinación) de tres trucos básicos:

1. Sacar un factor común a todos los términos.
2. Usar la fórmula de diferencia de cuadrados.
3. Hacer (y verificar) conjeturas educadas hasta encontrar la combinación correcta.

Los dos primeros trucos son totalmente mecánicos y requieren poco comentario. Aquí tienes un ejemplo que usa ambos trucos en turnos:

$2x^2 - 32 = 2(x^2 - 16) = 2(x - 4)(x + 4)$.

Nada de qué preocuparse: sacamos el $2$, notamos que uno de los factores resultantes era una diferencia de cuadrados, y así aplicamos la fórmula de diferencia de cuadrados. Deberías memorizar la fórmula de diferencia de cuadrados, ya que es muy útil.

El tercer truco, hacer conjeturas educadas, funciona mejor para polinomios de la forma $ax^2 + bx + c$. 

Por ejemplo, supongamos que queremos factorizar $x^2 + x - 12$. Bueno, si se puede factorizar, el resultado podría verse así:

$x^2 + x - 12 = (x)(x)$.

(Notemos que dejamos en blanco los términos constantes para mantener algo de flexibilidad).

Ten en cuenta que poner esos dos $x$ a la derecha fue una conjetura educada; lo hicimos porque su producto (el “F” de FOIL) es $x^2$, que coincide con uno de los términos de nuestro “polinomio objetivo” a la izquierda. Para continuar con nuestra conjetura educada, observamos que cualquiera que sean los términos constantes de los factores, su producto (el “L” de FOIL) debe ser $-12$ para coincidir con el término constante del polinomio objetivo. Hay muchas posibilidades aquí, así que elijamos un par de números cuyo producto sea $12$ y probémoslo: ¿Qué tal $(x - 4)(x + 3)$? ¿Servirá esto? Podemos comprobar multiplicándolo en nuestra cabeza; al hacerlo, encontramos que terminamos con el término $x$ incorrecto: terminamos con $-x$ en lugar de $+x$. Cerca, pero no hay cigarros. ¿Qué pasaría si simplemente cambiamos la ubicación del positivo y el negativo, así: $(x + 4)(x - 3)$? Una rápida comprobación muestra que esto sí funciona, así que nuestra factorización está completa:

$x^2 + x - 12 = (x + 4)(x - 3)$.

Y eso es todo lo que hay que hacer con el tercer truco. Con un poco de práctica, que harás en los ejercicios de abajo, desarrollarás una intuición para hacer buenas conjeturas.

Como puedes ver, factorizar eficientemente depende de poder multiplicar binomios en tu cabeza… que no es más que hacer FOIL… que no es más que usar la propiedad distributiva. Sí, requiere práctica y paciencia, pero todo aquí debería ser completamente comprensible.

### Ejercicios Seleccionados

Vamos a escoger unos ejercicios de aquí para resolver:

1. Factorizar $x^2 + x - 12$.
2. Factorizar $2x^2 - 32$.

Ahora, vamos a resolver esos problemas usando Python. Aquí tienes el código:



```python
import sympy as sp

# Definimos la variable
x = sp.symbols('x')

# Definimos los polinomios
polynomial_1 = x**2 + x - 12
polynomial_2 = 2*x**2 - 32

# Factorizamos los polinomios
factored_1 = sp.factor(polynomial_1)
factored_2 = sp.factor(polynomial_2)

print(f"Factorización de x^2 + x - 12: {factored_1}")
print(f"Factorización de 2x^2 - 32: {factored_2}")

```
