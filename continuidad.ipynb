### **2.9 Continuidad de funciones**

La **continuidad de una función** es cuando puedes mover una variable de forma suave, sin brincos ni huecos. Si una función no tiene baches, discontinuidades o saltos raros en su dominio, entonces decimos que es continua.

Para que una función $ f(x) $ sea continua en un punto $ a $, se deben cumplir tres condiciones:

1. **$ f(a) $ está definida**: El valor de la función en ese punto debe existir.
   
2. **El límite de $ f(x) $ conforme $ x \to a $ existe**: Conforme te acercas al punto $ a $, tanto por la izquierda como por la derecha, el límite debe converger a un valor.

3. **El valor del límite debe coincidir con $ f(a) $**: El valor al que llegas cuando te acercas a $ a $ debe ser el mismo que el valor de la función en $ a $. Es decir, no debe haber un brinco.

En resumen, para que la función sea continua en $ a $, se debe cumplir:
$$ \lim_{x \to a} f(x) = f(a) $$

### Ejemplo de una función continua

Por ejemplo, la función $ f(x) = x^2 $ es continua en todo su dominio porque no tiene brincos ni discontinuidades. Si te acercas a cualquier punto $ a $, el valor de la función es siempre suave.

### Ejemplo de una función discontinua

Ahora, considera la siguiente función:
$$
f(x) = 
\begin{cases} 
x^2 & \text{si } x < 1 \\
2 & \text{si } x = 1 \\
x + 1 & \text{si } x > 1
\end{cases}
$$

Aquí, la función tiene una discontinuidad en $ x = 1 $ porque el límite conforme $ x \to 1 $ desde la izquierda no es el mismo que el valor de $ f(1) $, lo que rompe la continuidad.

### Continuidad en intervalos

Si una función es continua en todos los puntos de un intervalo $ (a, b) $, decimos que es **continua en ese intervalo**. La gráfica de la función fluye sin interrupciones dentro de ese rango.

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función por tramos
def f(x):
    if x < 1:
        return x**2
    elif x == 1:
        return 2
    else:
        return x + 1

# Crear los valores de x en tres partes para evitar conectar en x=1
x_left = np.linspace(0, 1, 100, endpoint=False)  # No incluye x=1
x_right = np.linspace(1, 2, 100)  # Incluye x=1 en el segundo tramo

# Evaluar la función en cada parte
y_left = [f(xi) for xi in x_left]
y_right = [f(xi) for xi in x_right]

# Graficar los tramos
plt.plot(x_left, y_left, label="$x^2$ cuando $x<1$", color="blue")
plt.plot(x_right, y_right, label="$x+1$ cuando $x>1$", color="green")

# Agregar el valor de la función en x=1
plt.scatter(1, f(1), color="red", zorder=5, label="$f(1) = 2$")  # Punto donde f(1) = 2

# Marcamos la discontinuidad en x=1
plt.axvline(x=1, color="black", linestyle="--", label="x = 1 (Discontinuidad)")

# Configuraciones de la gráfica
plt.title("Función Discontinua")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.ylim(-1, 5)  # Ajustar límites para visualizar mejor
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función continua
def g(x):
    return x**2

# Crear valores de x
x = np.linspace(-1, 2, 100)

# Evaluar la función continua
y = g(x)

# Graficar la función continua
plt.plot(x, y, label="$g(x) = x^2$", color="blue")

# Punto donde g(1) = 1
plt.scatter(1, g(1), color="red", zorder=5, label="$g(1) = 1$")

# Configuraciones de la gráfica
plt.axvline(x=1, color="black", linestyle="--", label="x = 1")
plt.title("Función Continua")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.legend()
plt.grid(True)
plt.ylim(-1, 5)  # Ajustar límites para visualizar mejor
plt.show()
```
