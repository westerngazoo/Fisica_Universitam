## Definición Formal del Límite

El concepto de límite es fundamental en cálculo y análisis matemático. Nos permite definir cómo se comporta una función a medida que nos acercamos a un valor específico. Siguiendo la definición precisa de Spivak, decimos que el límite de \( f(x) \) cuando \( x \) tiende a \( a \) es \( L \), y lo denotamos como:

$$
\lim_{{x \to a}} f(x) = L
$$

si, para cualquier número positivo $( \epsilon )$, por pequeño que sea, existe un $( \delta > 0 )$ tal que siempre que $( 0 < |x - a| < \delta )$, se tiene que $( |f(x) - L| < \epsilon )$.

Esta definición formal también se conoce como la definición $( \epsilon )-( \delta )$.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos una función que tiene un límite claro en x = 2
def f(x):
    return (x**2 - 4)/(x - 2)

# Rango de valores para x (evitando la singularidad en x=2)
x_vals = np.linspace(1.5, 2.5, 400)
y_vals = f(x_vals)

# Graficamos la función
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"f(x) = $\frac{x^2 - 4}{x - 2}$")
plt.axvline(2, color='red', linestyle='--', label='x = 2')
plt.title("Límite cuando x tiende a 2")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend('')
plt.grid(True)
plt.show()
```

```python
from IPython.display import display, HTML
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Output, VBox, HBox, Label
import sympy as sp

# Aumentar el tamaño de la celda del notebook
display(HTML("<style>.jp-OutputArea { height: 800px; }</style>"))  # Aumenta la altura de la celda

# Definir la variable simbólica
x = sp.symbols('x')

# Función que tiene una indeterminación en x = a
def f(x):
    return (x**2 - 1**2) / (x - 1)

# Crear salida para los gráficos
output = Output()

# Inicializar posiciones de x
x_left = 0.99  # Comenzamos a la izquierda de 1
x_right = 1.01  # Comenzamos a la derecha de 1

# Crear sliders para los puntos
slider_left = FloatSlider(value=x_left, min=0.5, max=0.99, step=0.01, description='Punto Izquierda:')
slider_right = FloatSlider(value=x_right, min=1.01, max=2.0, step=0.01, description='Punto Derecha:')

# Gráfico inicial
x_vals = np.linspace(0, 2, 400)  # Valores de x
y_vals = f(x_vals)

# Configurar la figura y el gráfico inicial
plt.figure(figsize=(12, 10))  # Aumentar tamaño de la figura
plt.plot(x_vals, y_vals, label='f(x) = (x² - 1²) / (x - 1)', color='blue')
plt.axvline(1, color='red', linestyle='--', label='x = 1')

# Mostrar puntos iniciales
point_left, = plt.plot(slider_left.value, f(slider_left.value), 'go', zorder=5, label='Punto Izquierda')
point_right, = plt.plot(slider_right.value, f(slider_right.value), 'ro', zorder=5, label='Punto Derecha')

plt.title('Comportamiento de f(x) cerca de x = 1', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.ylim(-10, 10)  # Limitar el eje y para mejor visualización
plt.legend()
plt.grid(True)
plt.tight_layout()  # Ajusta automáticamente el tamaño del gráfico
plt.show()

# Texto para mostrar los valores
val_text_left = Label(value=f'Valor Izquierda: {f(slider_left.value)}')
val_text_right = Label(value=f'Valor Derecha: {f(slider_right.value)}')

# Función para actualizar los puntos y valores
def update_points(change=None):
    point_left.set_xdata(slider_left.value)
    point_left.set_ydata(f(slider_left.value))
    
    point_right.set_xdata(slider_right.value)
    point_right.set_ydata(f(slider_right.value))
    
    # Actualizar los valores en los labels
    val_text_left.value = f'Valor Izquierda: {f(slider_left.value)}'
    val_text_right.value = f'Valor Derecha: {f(slider_right.value)}'
    
    plt.draw()

# Calcular el límite simbólicamente
limit_result = sp.limit((x**2 - 1**2) / (x - 1), x, 1)

# Mostrar sliders, valores y resultado del límite
display(VBox([slider_left, val_text_left, slider_right, val_text_right, Label(value=f'Límite cuando x se aproxima a 1: {limit_result}')]), output)

# Asignar la función de actualización a los sliders
slider_left.observe(update_points, 'value')
slider_right.observe(update_points, 'value')
```

### Análisis del Límite de la Función

En esta sección, visualizamos la función $( f(x) = \frac{x^2 - 1^2}{x - 1} )$, que presenta una indeterminación en $( x = 1 )$. 

Utilizamos dos deslizadores para acercarnos al punto de indeterminación:

- **Deslizador "Punto Izquierda"**: Este deslizadores permite mover un punto hacia la izquierda de $( 1 )$. Al moverlo, podrás observar cómo el valor de $( f(x) )$ se acerca al límite.

- **Deslizador "Punto Derecha"**: Este deslizadores permite mover un punto hacia la derecha de $( 1 )$. De manera similar, podrás ver cómo $( f(x) )$ se aproxima al límite.

### Evaluación de la Función

Al lado de cada punto, se muestra el valor actual de $( f(x) )$ para que puedas ver cómo se comporta cuando te acercas al límite.

### Cálculo del Límite

El límite de la función cuando $( x )$ se acerca a $( 1 )$ se calcula como:

$$[
\lim_{x \to 1} f(x) = \lim_{x \to 1} \frac{x^2 - 1^2}{x - 1}
]$$

Al simplificar la función:

$$[
\lim_{x \to 1} f(x) = \lim_{x \to 1} \frac{(x - 1)(x + 1)}{x - 1} = \lim_{x \to 1} (x + 1) = 1 + 1 = 2
]$$

Así, al mover los puntos, podrás observar cómo se acerca el valor de \( f(x) \) al límite de \( 2 \) cuando \( x \) se aproxima a \( 1 \).


# Continuidad

Una función $f(x)$ se dice que es continua en un punto $c$ si se cumplen las siguientes condiciones:

1. $f(c)$ está definida.
2. $\lim_{x \to c} f(x)$ existe.
3. $\lim_{x \to c} f(x) = f(c)$.

### Tipos de Discontinuidades

- **Discontinuidad removible**: Si $\lim_{x \to c} f(x)$ existe pero $f(c)$ no está definida o es diferente a ese límite.
- **Discontinuidad de salto**: Si el límite desde la izquierda y el límite desde la derecha existen, pero son diferentes.
- **Discontinuidad infinita**: Si $f(x)$ tiende a infinito en $c$.

### Teoremas de Continuidad

- Teorema de la suma, producto y cociente: Si $f$ y $g$ son continuas en $c$, entonces $f + g$, $f \cdot g$ y $\frac{f}{g}$ (si $g(c) \neq 0$) también son continuas en $c$.


### Discontinuidad Removible

Vamos a graficar una función que tiene una discontinuidad removible. Usaremos la función:

$[
f(x) = \frac{x^2 - 1}{x - 1}
]$

Esta función tiene una discontinuidad removible en $( x = 1 )$ porque se puede simplificar a:

$[
f(x) = x + 1 \quad \text{(para } x \neq 1\text{)}
]$

Podemos redefinir $( f(1) = 2 )$ para hacer que la función sea continua. A continuación, mostraremos cómo se comporta la función alrededor de \( x = 1 \).


```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función con discontinuidad removible
def f(x):
    return (x**2 - 1) / (x - 1) if x != 1 else 2  # Definimos f(1) = 2

# Creamos un rango de valores x
x = np.linspace(0, 2, 100)
y = [f(i) for i in x]

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = \frac{x^2 - 1}{x - 1}$ con $f(1) = 2$", color='blue')

# Añadimos la discontinuidad
plt.scatter(1, f(1), color='red', zorder=5, label='Discontinuidad removible en $x=1$')
plt.axvline(1, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(2, color='gray', linestyle='--', linewidth=0.8)

# Configuración de la gráfica
plt.title('Ejemplo de Discontinuidad Removible')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.ylim(-1, 5)
plt.xlim(0, 2)
plt.legend()
plt.grid()
plt.show()
```
