# Tema 2: Límites y Continuidad

---

## 2.1 Definición de Límite

El límite de una función \( f(x) \) cuando \( x \) tiende a un valor \( c \) se denota como:

$$ \lim_{{x \to c}} f(x) $$

Significa que, conforme \( x \) se aproxima a \( c \), \( f(x) \) se acerca a un valor específico, llamado el **límite**. Esto se puede expresar como:

- Si \( f(x) \to L \) cuando \( x \to c \), entonces:

$$ \lim_{{x \to c}} f(x) = L $$

### Límite por la izquierda y por la derecha

- **Límite por la derecha** (cuando \( x \) se aproxima a \( c \) por valores mayores):

$$ \lim_{{x \to c^+}} f(x) $$

- **Límite por la izquierda** (cuando \( x \) se aproxima a \( c \) por valores menores):

$$ \lim_{{x \to c^-}} f(x) $$

Si ambos límites existen y son iguales, entonces el límite de \( f(x) \) en \( c \) existe.

---

## 2.2 Cálculo de Límites

Hay varias técnicas para calcular límites:

1. **Sustitución directa**: Si \( f(c) \) existe y es un valor finito, entonces:

$$ \lim_{{x \to c}} f(x) = f(c) $$

2. **Factorización**: Si la sustitución directa lleva a una indeterminación del tipo \( \frac{0}{0} \), podemos intentar factorizar y simplificar la expresión.

3. **Racionalización**: Para expresiones con raíces, podemos multiplicar por el conjugado para simplificar.

---

## 2.3 Continuidad de una Función

Una función \( f(x) \) es continua en un punto \( c \) si se cumplen las siguientes condiciones:

1. \( f(c) \) está definida.
2. El límite de \( f(x) \) cuando \( x \) tiende a \( c \) existe.
3. El valor de \( f(c) \) es igual al límite:

$$ \lim_{{x \to c}} f(x) = f(c) $$

Si la función es continua en todos los puntos de su dominio, decimos que \( f(x) \) es continua.

---

## Ejemplo en Python

Vamos a graficar una función que tiene una discontinuidad removible en \( x = 2 \).



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función con discontinuidad
def f(x):
    return np.where(x == 2, np.nan, (x**2 - 4) / (x - 2))

# Crear un rango de valores para x
x = np.linspace(1, 3, 400)
y = f(x)

# Graficar la función
plt.plot(x, y, label=r'$\frac{x^2 - 4}{x - 2}$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Función con Discontinuidad Removible')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

# Tema 2.3: Una Función que se Acerca al Infinito y Funciones Acotadas

---

## 2.3.1 Una Función que se Acerca al Infinito

Cuando una función \( f(x) \) se "acerca" al infinito, significa que su valor se hace muy grande o muy pequeño conforme \( x \) se aproxima a un cierto valor, ya sea finito o infinito. Esto puede ocurrir de dos formas:

1. **Límite hacia infinito**: Esto ocurre cuando el valor de la función crece sin límite conforme \( x \) tiende a cierto valor.
   
   Por ejemplo:

   $$ \lim_{{x \to 0}} \frac{1}{x} = \infty $$

   Esto significa que, conforme \( x \) se aproxima a 0 desde el lado positivo, \( f(x) = \frac{1}{x} \) crece indefinidamente.

2. **Límite en infinito**: Se refiere a cuando \( x \) tiende al infinito (o menos infinito) y queremos analizar el comportamiento de la función.

   Ejemplo:

   $$ \lim_{{x \to \infty}} \frac{1}{x} = 0 $$

   Aquí vemos que conforme \( x \) se va al infinito, el valor de la función \( f(x) = \frac{1}{x} \) se va acercando a 0.

---

## 2.3.2 Funciones Acotadas

Una función está **acotada** si existe un número que limita sus valores. En otras palabras, los valores de la función no pueden ser arbitrariamente grandes o pequeños, están restringidos dentro de un cierto rango.

- **Acotada superiormente**: Si existe un número \( M \) tal que para todo \( x \) en el dominio de \( f \), se cumple:

  $$ f(x) \leq M $$

- **Acotada inferiormente**: Si existe un número \( m \) tal que para todo \( x \), se cumple:

  $$ f(x) \geq m $$

- **Acotada**: Una función es acotada si es acotada superior e inferiormente, es decir, si existen \( M \) y \( m \) tales que:

  $$ m \leq f(x) \leq M $$

### Ejemplos

- La función \( f(x) = \sin(x) \) es acotada porque sus valores están siempre entre -1 y 1:

  $$ -1 \leq \sin(x) \leq 1 $$

- La función \( f(x) = \frac{1}{x} \) no está acotada en \( x = 0 \), porque sus valores se van al infinito cuando \( x \) tiende a 0.

---

## Ejemplo en Python

Vamos a graficar una función que se acerca al infinito en \( x = 0 \) y otra que es acotada, como el \( \sin(x) \).




```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f_infinito(x):
    return 1/x

def f_acotada(x):
    return np.sin(x)

# Crear un rango de valores para x
x1 = np.linspace(0.1, 5, 400)  # Evitar el 0 en 1/x para evitar división por cero
x2 = np.linspace(-2*np.pi, 2*np.pi, 400)

# Graficar la función que se acerca al infinito
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x1, f_infinito(x1), label=r'$\frac{1}{x}$')
plt.title('Función que se Acerca al Infinito : '+r'$\frac{1}{x}$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Graficar la función acotada
plt.subplot(1, 2, 2)
plt.plot(x2, f_acotada(x2), label=r'$\sin(x)$', color='orange')
plt.title('Función Acotada: $\sin(x)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
