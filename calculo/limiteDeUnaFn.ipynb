# Tema 2.6: El Límite de la Función $( \frac{\sin x}{x} )$ cuando $( x \to 0 )$

---

## Definición:

Uno de los límites más importantes en cálculo es:

$$ \lim_{{x \to 0}} \frac{\sin x}{x} = 1 $$

Este límite aparece frecuentemente en el cálculo diferencial y se puede demostrar usando diversas técnicas, incluyendo la aproximación mediante series de Taylor o utilizando el teorema del sándwich.

---

### Demostración usando el Teorema del Sándwich

Sabemos que para \( x \) cercano a 0:

$$ \cos x \leq \frac{\sin x}{x} \leq 1 $$

Tomando los límites de ambos lados cuando $( x \to 0 )$, obtenemos que el límite de $( \frac{\sin x}{x} )$ también es 1.

---

### Ejemplo en Python

Vamos a visualizar cómo $( \frac{\sin x}{x} )$ se comporta cuando $( x \to 0 )$.



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def sinc(x):
    return np.sinc(x/np.pi)  # np.sinc está normalizada como sin(πx)/(πx)

# Crear un rango de valores para x
x = np.linspace(-10, 10, 400)

# Graficar
plt.plot(x, sinc(x), label=r'$\frac{\sin(x)}{x}$')
plt.title(r'Gráfico de $\frac{\sin(x)}{x}$ cuando $x \to 0$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

# Límite simbólico de $( \frac{\sin x}{x} )$ cuando $( x \to 0 )$

Ahora vamos a calcular el límite de manera simbólica utilizando `SymPy`.

---

### Ejemplo en Python utilizando SymPy:



```python
import sympy as sp

# Definir la variable simbólica
x = sp.symbols('x')

# Definir la función
funcion = sp.sin(x) / x

# Calcular el límite
limite = sp.limit(funcion, x, 0)

# Mostrar el resultado
limite
```
