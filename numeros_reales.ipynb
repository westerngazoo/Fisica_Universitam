## 1.1 Conjuntos de Números Reales

Los números reales \( \mathbb{R} \) son aquellos que incluyen tanto los racionales como los irracionales. Son los números que pueden ubicarse en la recta numérica. Entre ellos se encuentran:

- **Naturales**: \( \mathbb{N} = \{1, 2, 3, \dots\} \)
- **Enteros**: \( \mathbb{Z} = \{\dots, -2, -1, 0, 1, 2, \dots\} \)
- **Racionales**: $$ \mathbb{Q} = \left\{ \frac{p}{q} \mid p \in \mathbb{Z}, q \in \mathbb{Z}, q \neq 0 \right\} $$
- **Irracionales**: Números que no pueden expresarse como cociente de enteros, como $$ \sqrt{2} $$ ó $$ \pi $$.

---

## 1.2 Propiedades de los Números Reales

Los números reales tienen varias propiedades importantes:

1. **Propiedad conmutativa de la suma**:
    $$ a + b = b + a $$

2. **Propiedad conmutativa del producto**:
    $$ ab = ba $$

3. **Propiedad distributiva**:
    $$ a(b + c) = ab + ac $$

---

## 1.3 Funciones

Una función \( f \) es una relación entre dos conjuntos, tal que a cada elemento del primer conjunto le corresponde uno y solo un elemento del segundo conjunto. Se denota como:

$$ f: A \to B $$

Donde \( A \) es el **dominio** y \( B \) es el **codominio**. Si \( x \in A \), entonces \( f(x) \in B \).

---

## 1.4 Función Lineal

Una función lineal es una función de la forma:

$$ f(x) = mx + b $$

Donde \( m \) es la pendiente y \( b \) es la intersección con el eje \( y \).

---

## Ejemplo en Python

Vamos a graficar una función lineal con \( m = 2 \) y \( b = 3 \).



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función lineal
def f(x):
    return 2*x + 3

# Crear un rango de valores para x
x = np.linspace(-10, 10, 400)
y = f(x)

# Graficar la función
plt.plot(x, y, label=r'$f(x) = 2x + 3$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Función Lineal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
