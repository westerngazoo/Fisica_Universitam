### 2.10 Ciertas propiedades de las funciones continuas

#### Definición de Continuidad

Una función $f(x)$ se dice que es continua en un punto $a$ si se cumplen las siguientes tres condiciones:

1. $f(a)$ está definida.
2. El límite de $f(x)$ cuando $x$ se aproxima a $a$ existe:
   $$ \lim_{x \to a} f(x) \text{ existe.} $$
3. El límite de $f(x)$ cuando $x$ se aproxima a $a$ es igual al valor de la función en ese punto:
   $$ \lim_{x \to a} f(x) = f(a). $$

Si una función cumple estas condiciones en todos los puntos de su dominio, se dice que es continua en su dominio.

#### Propiedades de Funciones Continuas

1. **Suma de Funciones Continuas**: Si $f(x)$ y $g(x)$ son funciones continuas en un punto $a$, entonces la suma $f(x) + g(x)$ también es continua en $a$:
   $$ \lim_{x \to a} \left(f(x) + g(x)\right) = \lim_{x \to a} f(x) + \lim_{x \to a} g(x). $$

2. **Producto de Funciones Continuas**: Si $f(x)$ y $g(x)$ son funciones continuas en un punto $a$, entonces el producto $f(x) \cdot g(x)$ también es continuo en $a$:
   $$ \lim_{x \to a} \left(f(x) \cdot g(x)\right) = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x). $$

3. **Cociente de Funciones Continuas**: Si $f(x)$ y $g(x)$ son funciones continuas en un punto $a$, y $g(a) \neq 0$, entonces el cociente $f(x) / g(x)$ también es continuo en $a$:
   $$ \lim_{x \to a} \left(\frac{f(x)}{g(x)}\right) = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}. $$

4. **Composición de Funciones Continuas**: Si $f(x)$ es continua en un punto $b$ y $g(x)$ es continua en $a$, y si $f(g(x))$ está definida, entonces la composición $f(g(x))$ también es continua en $a$:
   $$ \lim_{x \to a} f(g(x)) = f\left(\lim_{x \to a} g(x)\right). $$

#### Ejemplo Práctico en Python



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función continua f(x) = x^2
def f(x):
    return x**2

# Definir los valores de x
x = np.linspace(-2, 2, 400)
y = f(x)

# Graficar la función continua
plt.plot(x, y, label="$f(x) = x^2$", color="blue")

# Configuraciones de la gráfica
plt.title("Función Continua")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
