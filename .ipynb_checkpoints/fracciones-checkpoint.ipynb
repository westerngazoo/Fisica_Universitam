# Fracciones: Dos reglas intuitivas que llevan a todas las demás

"Cualquier tonto puede saber. El punto es entender."  
- Albert Einstein

Comenzaremos nuestro estudio de fracciones con una observación: Obtienes la misma cantidad de postre al tomar dos quintos de un pie que al tomar un quinto dos veces. O, traduciendo esta declaración en símbolos,  
$$\frac{2}{5} = 2 \left(\frac{1}{5}\right).$$

Por supuesto, no había nada especial en $2$ o $5$ en ese último ejemplo. También podríamos haber notado que siete octavos de un pie es lo mismo que un octavo tomado siete veces, así que  
$$\frac{7}{8} = 7\left(\frac{1}{8}\right).$$  

El álgebra, la ciencia de los patrones, nos permite describir el patrón que estamos viendo aquí de la siguiente manera.

Ahora, para una segunda observación. Dado que $10$ décimos hacen un entero, cortar cada décimo en tercios nos daría $30$ partes iguales. Así, cada tercio de un décimo es exactamente $\frac{1}{30}$ del todo. O, en símbolos,  
$$\left(\frac{1}{3}\right) \left(\frac{1}{10}\right) = \frac{1}{30}.$$

De manera similar, un octavo de un medio debe ser un dieciseisavo, así que  
$$\left(\frac{1}{8}\right)\left(\frac{1}{2}\right) = \frac{1}{16}.$$  

El patrón algebraico aquí es:  
Tomaremos como un axioma que nuestras dos "reglas intuitivas de fracciones" se mantienen para todos los valores de $a$ y $b$.  
## Regla Intuitiva de Fracción 1:
$$\frac{a}{b} = a \left(\frac{1}{b}\right)$$

## Regla Intuitiva de Fracción 2:
$$\left(\frac{1}{a}\right) \left(\frac{1}{b}\right) = \frac{1}{ab}$$


Ahora, para una sorpresa: En las próximas páginas, veremos que cada aspecto de la aritmética fraccionaria – ese tema que confunde a tantas personas – sigue lógicamente de esas dos reglas simples. ¡Esto debería alentar a quienes encuentran confusas las fracciones! A lo largo del camino, también aclararemos bastante álgebra.

## Multiplicación y Reducción de Fracciones

A partir de las dos reglas intuitivas de fracción, podemos explicar completamente la regla bien conocida para multiplicar fracciones.  
(Y si has resuelto el ejercicio 24, ya lo has descubierto por ti mismo.)

### Prueba
$$\frac{a}{b} \cdot \frac{c}{d} = a \left(\frac{1}{b}\right) c \left(\frac{1}{d}\right) \quad \text{(aplicando la Regla Intuitiva de Fracción 1)}$$  
$$= a c \left(\frac{1}{b}\right) \left(\frac{1}{d}\right) \quad \text{(reordenando la multiplicación)}$$  
$$= a c \left(\frac{1}{bd}\right) \quad \text{(Regla Intuitiva de Fracción 2)}$$  
$$= \frac{ac}{bd} \quad \text{(Regla Intuitiva de Fracción 1)} \; \square$$

Así, por ejemplo,  
$$\left(\frac{6}{7}\right)\left(\frac{2}{3}\right) = \frac{12}{21}.$$  
Probablemente también sepas que $$\frac{12}{21}$$ se puede reducir a $$\frac{4}{7}$$. Pero, ¿puedes explicar por qué se nos permite eliminar ese factor común de 3 de su numerador y denominador? Esta no es una pregunta ociosa. Las personas que no entienden por qué las fracciones numéricas pueden ser reducidas (incluso si saben cómo hacerlo) tienden a equivocarse en la operación algebraica análoga, que discutiremos en la siguiente sección.

Reducimos una fracción eliminando factores que son comunes a su numerador y denominador. Un ejemplo mostrará exactamente por qué podemos hacer esto:

$$\frac{12}{21} = \frac{4 \cdot 3}{7 \cdot 3} = \frac{4}{7} \cdot \frac{3}{3} = \frac{4}{7} \cdot 1 = \frac{4}{7}.$$

Por lo tanto, reducir una fracción no es más que eliminar un factor oculto de 1. Muy simple.


## Regla de Multiplicación para Fracciones

Para multiplicar fracciones, multiplicamos los numeradores y multiplicamos los denominadores. O en símbolos:

$$\frac{a}{b} \cdot \frac{c}{d} = \frac{ac}{bd}$$



## Cancelando Arriba y Abajo de la Barra

*Y que no haya quejas de la barra,  
Cuando me adentre en el mar.*  
- Alfred Lord Tennyson, *Crossing the Bar*

Todo el mundo sabe que podemos simplificar $ \frac{ab + a}{a} $ al cancelar algunos $ a $, pero no todo el mundo sabe por qué. En consecuencia, muchos estudiantes producen "simplificaciones" incorrectas como $ ab $ o $ b + a $. Aquellos que realmente entienden álgebra nunca cometen tales errores, pues saben que cancelar es simplemente descartar un factor oculto de 1. La lógica aquí es la misma que con la reducción de fracciones numéricas ordinarias.

**Ejemplo:** Simplifica $ \frac{ab + a}{a} $.

**Solución:**

$$ 
\frac{ab + a}{a} = \frac{a(b + 1)}{a} \quad (\text{factoreando } a \text{ del numerador}) 
$$

$$ 
= \frac{a}{a} \cdot \frac{b + 1}{1} \quad (\text{por la regla de multiplicación para fracciones}) 
$$

$$ 
= b + 1. \quad (\text{porque } \frac{a}{a} = 1) \, \blacksquare 
$$

Repito: “Cancelar arriba y abajo de la barra” es simplemente una forma abreviada de descartar un factor oculto de 1. Es esencial que entiendas esto a fondo. Siempre que te sientas tentado a "cancelar" algo de arriba y abajo de una barra de fracción, pregúntate si puedes separarlo como un factor de 1. Si puedes, puedes cancelar; si no, no puedes cancelar. Eso es todo.


## El Viejo Truco de Multiplicar por 1

Para reducir una fracción, eliminamos un factor de 1. Sorprendentemente, introducir un factor de 1 también puede ser útil. Multiplicar una fracción por un 1 disfrazado de manera conveniente preserva el valor de la fracción pero cambia su forma a algo más útil. Yo llamo a esto “el viejo truco de multiplicar por 1”. Captura muy bien el espíritu del álgebra.*

Aquí hay un ejemplo fácil de cómo funciona: $ \frac{3}{5} $ es cuántos vigésimos? Para responder a esta pregunta, observamos que multiplicar el fondo de nuestra fracción por 4 lo cambiaría a 20, que es justo lo que queremos. Buenas noticias: podemos obtener lo que queremos si también multiplicamos la parte de arriba por 4, porque el efecto neto será multiplicar la fracción por 1, lo que preserva su valor. Así:

$$ 
\frac{3}{5} = \frac{3}{5} \cdot \frac{4}{4} = \frac{12}{20} 
$$

Tal es el viejo truco de multiplicar por 1. Ahora veamos esto en una aplicación más sustancial.



## Regla de División de Fracciones

**Para dividir una fracción por otra, "invertimos y multiplicamos".** O en símbolos,

$$\frac{\frac{a}{b}}{\frac{c}{d}}=\frac{a}{b}\cdot\frac{d}{c}$$

**Demostración:**

$$\frac{\frac{a}{b}}{\frac{c}{d}}=\frac{\frac{a}{b}}{\frac{c}{d}}\cdot\frac{\frac{d}{c}}{\frac{d}{c}}$$ (el viejo truco de multiplicar por 1)

$$=\frac{\frac{ad}{bc}}{1}$$ (regla de multiplicación de fracciones)

$$=\frac{ad}{bc}$$ (dividir por 1 no hace nada)

$$=\frac{a}{b}\cdot\frac{d}{c}$$ (regla de multiplicación de fracciones)

**Después de haber demostrado el teorema, consideremos un ejemplo algebraico típico en el que podemos aplicarlo.**

*Hay un truco análogo de "sumar 0", que usamos, por ejemplo, al completar el cuadrado, un truco que encontrarás en el Capítulo 3.

**Un buen nombre para un libro de álgebra sería "El arte de sumar cero y multiplicar por uno".**

## Sumando y Restando Fracciones

**Para sumar o restar fracciones con el mismo denominador, simplemente sumamos o restamos los numeradores y mantenemos el mismo denominador.**

**Si las fracciones tienen distinto denominador, necesitamos encontrar un denominador común.** Este es el número más pequeño que ambos denominadores dividen exactamente. Luego, convertimos cada fracción a una equivalente con este nuevo denominador y realizamos la operación.

**El "truco de multiplicar por 1"** nos permite convertir fracciones a equivalentes con un denominador común. Multiplicamos el numerador y el denominador de cada fracción por el número necesario para obtener el denominador común.

### Ejemplo Numérico:

$ \frac{5}{6} - \frac{1}{10} $

1.  **Encuentra el mínimo común múltiplo (MCM):** En este caso, el MCM de 6 y 10 es 30.
2.  **Convierte las fracciones:**
    * $ \frac{5}{6} \cdot \frac{5}{5} = \frac{25}{30} $
    * $ \frac{1}{10} \cdot \frac{3}{3} = \frac{3}{30} $
3.  **Resta:** $ \frac{25}{30} - \frac{3}{30} = \frac{22}{30} = \frac{11}{15} $

### Ejemplo Algebraico:

$ \frac{14d^2 + 2d + c}{c\sqrt{b}} - \frac{3d - c}{c\sqrt{b}} $

1.  **Mismo denominador:** ¡Ya tenemos el mismo denominador!
2.  **Resta los numeradores:** $(14d^2 + 2d + c) - (3d - c)$
3.  **Simplifica:** $14d^2 + 2d + c - 3d + c = 14d^2 - d + 2c$
4.  **Resultado final:** $ \frac{14d^2 - d + 2c}{c\sqrt{b}} $

**¿Por qué los paréntesis?**
Cuando restamos expresiones algebraicas, es crucial usar paréntesis para indicar que estamos restando toda la expresión, no solo el primer término. Esto evita errores comunes.

### Resumen

* **Siempre busca el MCM:** Simplifica los cálculos y hace más fácil encontrar el resultado final.
* **Los paréntesis son clave:** En expresiones algebraicas, los paréntesis garantizan que se reste toda la expresión.
* **Simplifica al final:** Una vez que hayas realizado la suma o resta, simplifica la fracción resultante si es posible.
